```python
from sentence_transformers import SentenceTransformer
sentences = ["Cô giáo đang ăn kem", "Chị gái đang thử món thịt dê"]

model = SentenceTransformer('keepitreal/vietnamese-sbert')
embeddings = model.encode(sentences)
print(embeddings)
```

```python
import os 
import glob
import numpy as np

################# Config Folder #################
root_path = '/content/drive/MyDrive/ASR_Vietnamese_T'
txt_paths = glob.glob('/content/drive/MyDrive/ASR_Vietnamese/TRANSCRIPTC02_V00_batch_2/*/*.txt')
txt_paths = sorted(txt_paths)
#################################################

for txt_path in txt_paths:
  batch_name = txt_path.split('/')[-3].replace('TRANSCRIPT','')  
  video_name = txt_path.split('/')[-2]

  # if int(video_name.split('_')[-1].replace('V','')) <= 164:
  #   continue
  
  des_path = f'{root_path}/Embed{batch_name}/{video_name}'
  os.makedirs(des_path, exist_ok=True)

  with open(txt_path, 'rb') as f:
    line = f.read()
  
  line = line.decode('utf-8')
  text = [line, ]
  
  if text[0] == '':
    print(f'Skip: {txt_path}')
    continue

  else:
    text_embeddings = model.encode(text)
    name_npy = txt_path.split('/')[-1].replace('.txt', '')
    outfile = f'{des_path}/{name_npy}.npy'
    
    np.save(outfile, text_embeddings)
```
